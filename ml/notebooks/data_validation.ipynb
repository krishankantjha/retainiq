# Data Validation - Telco Customer Churn

This notebook checks the dataset before EDA and modeling.

- **Dataset:** IBM Telco Customer Churn
- **Target column:** Churn

## Load Data

```python
import pandas as pd
import numpy as np
import logging
from pathlib import Path

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)
logger = logging.getLogger("data_validation")

df = pd.read_csv(Path("../../data/raw/Telco_Customer_Churn.csv"))
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
logger.info(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

## 1. Dataset Shape Validation

```python
EXPECTED_ROWS = 7043
EXPECTED_COLS = 21

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

assert df.shape[0] == EXPECTED_ROWS, f"Expected {EXPECTED_ROWS} rows but got {df.shape[0]}"
assert df.shape[1] == EXPECTED_COLS, f"Expected {EXPECTED_COLS} columns but got {df.shape[1]}"

logger.info(f"Shape check passed: {df.shape[0]} rows, {df.shape[1]} columns")
```

**Note:** We have data to analyze. Empty file would be a problem.

## 2. Schema Validation

```python
EXPECTED_COLUMNS = [
    "customerID", "gender", "SeniorCitizen", "Partner", "Dependents",
    "tenure", "PhoneService", "MultipleLines", "InternetService",
    "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport",
    "StreamingTV", "StreamingMovies", "Contract", "PaperlessBilling",
    "PaymentMethod", "MonthlyCharges", "TotalCharges", "Churn"
]

print(df.columns.tolist())
print("\nNumber of columns:", len(df.columns))

missing = set(EXPECTED_COLUMNS) - set(df.columns)
extra = set(df.columns) - set(EXPECTED_COLUMNS)

assert len(missing) == 0, f"Missing columns: {missing}"
assert len(extra) == 0, f"Unexpected extra columns: {extra}"

logger.info("Schema check passed: all 21 expected columns present with correct names")
```

**Note:** Churn column must be there because it is our target.

**Note:** TotalCharges should be converted to number in data cleaning step.

## 4. Missing Value Validation

```python
for col in df.select_dtypes(include="object").columns:
    blank_count = (df[col].astype(str).str.strip() == "").sum()
    if blank_count > 0:
        logger.warning(f"{col} has {blank_count} blank/whitespace values")

null_counts = df.isnull().sum()
null_counts = null_counts[null_counts > 0]
if len(null_counts) > 0:
    logger.warning(f"Null values found:\n{null_counts}")
else:
    logger.info("Null check passed: no NaN values found")

print("\nNull check (NaN only):")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
(df.isnull().sum() / len(df) * 100).round(2)
```

```python
blank_total = (df["TotalCharges"].astype(str).str.strip() == "").sum()
print("Blank TotalCharges:", blank_total)
print("Missing after numeric conversion:", df["TotalCharges"].isnull().sum())
```

**Note:** Some new customers have blank TotalCharges when tenure is 0.

## 5. Duplicate Record Validation

```python
assert df.duplicated().sum() == 0, "Duplicate rows found"
assert df["customerID"].duplicated().sum() == 0, "Duplicate customerID found"
logger.info("Duplicate check passed: no duplicate rows or customerIDs")
```

**Note:** No duplicates is good - each row is one customer.

## 7. Numerical Data Validation

```python
df[["SeniorCitizen", "tenure", "MonthlyCharges", "TotalCharges"]].describe()
```

```python
assert (df["tenure"] < 0).sum() == 0, "Negative tenure values found"
assert (df["MonthlyCharges"] < 0).sum() == 0, "Negative MonthlyCharges values found"
assert (df["TotalCharges"] < 0).sum() == 0, "Negative TotalCharges values found"

print("Negative tenure: 0")
print("Negative MonthlyCharges: 0")
print("Negative TotalCharges: 0")

for col in ["tenure", "MonthlyCharges", "TotalCharges"]:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col} outliers: {len(outliers)}")
```

**Note:** Values look reasonable. Tenure is between 0 and 72 months.

## 8. Categorical Data Validation

```python
expected_categories = {
    "gender": {"Female", "Male"},
    "Partner": {"Yes", "No"},
    "Dependents": {"Yes", "No"},
    "PhoneService": {"Yes", "No"},
    "MultipleLines": {"No phone service", "No", "Yes"},
    "InternetService": {"DSL", "Fiber optic", "No"},
    "OnlineSecurity": {"No", "Yes", "No internet service"},
    "OnlineBackup": {"Yes", "No", "No internet service"},
    "DeviceProtection": {"No", "Yes", "No internet service"},
    "TechSupport": {"No", "Yes", "No internet service"},
    "StreamingTV": {"No", "Yes", "No internet service"},
    "StreamingMovies": {"No", "Yes", "No internet service"},
    "Contract": {"Month-to-month", "One year", "Two year"},
    "PaperlessBilling": {"Yes", "No"},
    "PaymentMethod": {
        "Electronic check",
        "Mailed check",
        "Bank transfer (automatic)",
        "Credit card (automatic)",
    },
    "Churn": {"No", "Yes"},
}

for col, expected in expected_categories.items():
    actual = set(df[col].unique())
    assert actual.issubset(expected), f"Unexpected values in {col}: {actual - expected}"
    print(f"{col}: {df[col].unique()}")
```

**Note:** Categories look normal - Yes/No, contract types, payment methods etc.

## 10. Data Leakage Check

```python
forbidden_cols = ["customerID", "Churn"]
features = [col for col in df.columns if col not in forbidden_cols]

for col in forbidden_cols:
    assert col not in features, f"Data leakage error: {col} is in the features list!"

logger.info(f"Leakage check passed: {len(features)} features selected, ID and Target excluded")
```

```python
corr_tenure = df["TotalCharges"].corr(df["tenure"])
corr_monthly = df["TotalCharges"].corr(df["MonthlyCharges"])

print(f"Correlation between TotalCharges and tenure: {round(corr_tenure, 4)}")
print(f"Correlation between TotalCharges and MonthlyCharges: {round(corr_monthly, 4)}")

if corr_tenure > 0.8 or corr_monthly > 0.8:
    logger.warning("TotalCharges is highly correlated with another feature. Consider dropping it during feature engineering.")
else:
    logger.info("TotalCharges multicollinearity check passed.")
```

**Note:** Using target as feature would cheat the model. ID has no meaning for prediction.

## 11. Business Rule Validation

```python
print("Tenure = 0 customers:", (df["tenure"] == 0).sum())

tenure_zero = df[df["tenure"] == 0]
blank = (tenure_zero["TotalCharges"].astype(str).str.strip() == "").sum()
print("Blank TotalCharges when tenure is 0:", blank)
```

```python
mask = (df["tenure"] >= 1) & df["TotalCharges"].notna()
issue = df[mask & (df["TotalCharges"] < df["MonthlyCharges"])]
assert len(issue) == 0, f"Business rule violation: {len(issue)} customers have TotalCharges < MonthlyCharges"
logger.info("Business rule check passed: no TotalCharges < MonthlyCharges violations")
```

```python
internet_cols = ["OnlineSecurity", "OnlineBackup", "DeviceProtection",
                 "TechSupport", "StreamingTV", "StreamingMovies"]

for col in internet_cols:
    violations = df[(df["InternetService"] == "No") & (df[col] != "No internet service")]
    assert len(violations) == 0, f"{col} has {len(violations)} rows inconsistent with InternetService=No"

phone_violations = df[(df["PhoneService"] == "No") & (df["MultipleLines"] != "No phone service")]
assert len(phone_violations) == 0, f"MultipleLines has {len(phone_violations)} rows inconsistent with PhoneService=No"

logger.info("Cross-column consistency check passed!")
```

**Note:** New customers (tenure 0) may not have total charges yet. That is normal.

## 12. Validation Summary
